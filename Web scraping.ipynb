```python
%%time
# noob1 implementation
# search for Covid 19
# load the webpage
result = requests.get("https://www.whitehouse.gov/briefings-statements/")
# specify keywords to serach the website
queries = ["Covid19","covid","Covid","corona","Corona","COVID19","COVID", "COVID-19"]
# create a list to store all the related links
urls = []
# counter for page number
page_number = 1
# if result has some value: run 
while(result):
    # flag to check whether if next page is available
    next_page_availability = 0
    # create a soup object for current page
    page = BeautifulSoup(result.content, 'lxml')
    # collect all the h2 tags, bcz debriefs are listed under these tags when we inspected the website 
    h_tags = page.find_all("h2")
    # loop through all the tags and find query related debrief titles
    for h_tag in h_tags:
        a_tag = h_tag.find("a")
        for i in queries:
            if i in a_tag.text:
                print("yes")
                urls.append(a_tag.attrs["href"])
                break
    #The below code is to find next page's link
    a_tags = page.find_all('a')
    for a_tag in a_tags:
        if "class" in a_tag.attrs:
            if a_tag.attrs["class"][0] == "pagination__next":
                result = requests.get(a_tag.attrs["href"])
                next_page_availability += 1
    # the below code is to check if there's no next page 
    if next_page_availability == 0:
        print("completed")
        break
    # printing statement for understanding
    print("completed page number = ", page_number)
    page_number+=1
```

```python
len(urls)
```

```python
%%time
# noob2 implementation
# search for President Donald Trump
result = requests.get("https://www.whitehouse.gov/briefings-statements/")
queries = ["Trump","trump","President","president"]
urls = []
page_number = 1
while(result):
    next_page_availability = 0
    page = BeautifulSoup(result.content, 'lxml')
    h_tags = page.find_all("h2")
    for h_tag in h_tags:
        a_tag = h_tag.find("a")
        for i in queries:
            if i in a_tag.text:
                print("yes")
                urls.append(a_tag.attrs["href"])
                break
    # find all the elements with a tag "and" class attribute of that tag is equal to ['pagination__next'].
    a_tags = page.find_all('a', class_ = ['pagination__next']) # we need to add _ after attribute name like "class_" or "href_" inside find_all function
    if a_tags[0]:
        result = requests.get(a_tags[0]['href'])
        next_page_availability += 1
    if next_page_availability == 0:
        print("completed")
        break
    print("completed page number = ", page_number)
    page_number+=1
```
